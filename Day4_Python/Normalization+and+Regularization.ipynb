## l1-Norm, l2-Norm

**What is the norm?**

Mathematically a norm is a total size or length of all vectors in a vector space  or matrices. For simplicity, we can say that the higher the norm is, the bigger the (value in) matrix or vector is

General formula is : $||x||_p \ = \sqrt[n]{x^p_1+x^p_2+...+x^p_n}$

$||x||_2 \ = \sqrt{x^2_1+x^2_2+...+x^2_n}$

$||x||_1 \ = |x_1|+|x_2|+...+|x_n|$

** Let us try in python **



```python
import numpy as np
a = np.array([1, 2])
L2=np.linalg.norm(a,ord=2)
L1=np.linalg.norm(a,ord=1)
print(L2,L1)
```

```python
import numpy as np
a = np.array([10000, 20000])
L2=np.linalg.norm(a,ord=2)
L1=np.linalg.norm(a,ord=1)
print(L2,L1)
```

## l1-Norm, l2-Norm and Normalization

Normalization is the process of scaling individual samples to have unit norm.

See example

```python
from sklearn import preprocessing
import numpy as np

X = [[ 1., -1.,  2.],
     [ 2.,  0.,  0.],
     [ 0.,  1., -1.]]
X_normalized = preprocessing.normalize(X, norm='l2')
print(X_normalized)
```
